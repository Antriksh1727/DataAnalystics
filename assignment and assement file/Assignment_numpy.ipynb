```python
#18.  Describe a practical example where NumPy can be used in EDA, AI, ML, and DL, and implement a NumPy solution for a simple task in each 
#area. 

#EDA - You have survey data with some missing values, and you want to replace them with the mean

# Survey age data with missing values (NaN)
ages = np.array([25, 30, np.nan, 22, 28, np.nan, 35])

# Find mean without NaN
mean_age = np.nanmean(ages)

# Replace NaN with mean
ages_cleaned = np.where(np.isnan(ages), mean_age, ages)

print("Cleaned ages:", ages_cleaned)

# AI - A basic AI system to recommend “Hot” or “Cold” drink based on temperature

temperature_data = np.array([15, 32, 28, 10, 35])  # in °C

# AI decision: if temp >= 25°C → "Cold drink", else → "Hot drink"
recommendation = np.where(temperature_data >= 25, "Cold drink", "Hot drink")

print("Recommendations:", recommendation)

# ML - Scaling features between 0 and 1 before feeding them to a model.

features = np.array([[50, 2000],
                     [20, 1200],
                     [30, 1500]])

# Min-Max normalization
min_vals = features.min(axis=0)
max_vals = features.max(axis=0)
normalized_features = (features - min_vals) / (max_vals - min_vals)

print("Normalized features:\n", normalized_features)

# DL - Converting integer class labels into one-hot vectors (common in neural networks).

labels = np.array([0, 2, 1, 2])  # Class IDs
num_classes = 3

# One-hot encoding
one_hot_labels = np.eye(num_classes)[labels]

print("One-hot labels:\n", one_hot_labels)



```

```python
#19. Create a random 4x4 matrix and use NumPy to compute its eigenvalues and eigenvectors. Verify the eigenvalues by reconstructing the matrix. 


# Create a random 4x4 matrix
A = np.random.rand(4, 4)

print("Original Matrix A:\n", A)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)

# Reconstruct the matrix: A = V * D * V_inv
D = np.diag(eigenvalues)            # Diagonal matrix of eigenvalues
V = eigenvectors
V_inv = np.linalg.inv(V)

A_reconstructed = V @ D @ V_inv

print("\nReconstructed Matrix A:\n", A_reconstructed)

# Verify (allowing for small floating-point errors)
print("\nDifference between A and A_reconstructed:\n", A - A_reconstructed)
```

```python
#21. : Perform matrix multiplication of two 2D arrays using np.dot () and @. Compare the results and performance of both methods using a large dataset.

import time

A = np.random.rand(1000, 500)
B = np.random.rand(500, 800)

#using dot
start_dot = time.time()
result_dot = np.dot(A, B)
end_dot = time.time()

#using @
start_at = time.time()
result_at = A @ B
end_at = time.time()

#compare result
print("Are results identical?", np.allclose(result_dot, result_at))
print()

# Compare performance
print(f"np.dot() time: {end_dot - start_dot:.6f} seconds")
print(f"@ operator time: {end_at - start_at:.6f} seconds")
```

```python
#22.  Create a 3D array of shape (2, 1, 4) and a 2D array of shape (4, 1). Perform an element-wise operation using broadcasting and explain the result. Use np. new axis to achieve the same result without broadcasting. 

a = np.arange(8).reshape(2,1,4)
b = np.arange(4).reshape(4,1)

print("3D array shape:", a.shape)
print("2D array (reshaped to 1D) shape:", b.shape)

#broadcast
result = a+b  
print("\nResult with broadcasting:\n", result)
print("Shape after broadcasting:", result.shape)

#using np.newaxis
b_new = b[np.newaxis, np.newaxis, :]  # shape (1, 1, 4)
result_newaxis = a + b_new
print("\nResult with np.newaxis:\n", result_newaxis)
print("Shape after np.newaxis:", result_newaxis.shape)
```
