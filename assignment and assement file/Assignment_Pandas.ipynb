```python
import numpy as np
import pandas as pd
```

```python
#1. Create a series of three different colors 

colors = ["Red","Green","Black"]

color_series = pd.Series(colors)

print(color_series)
```

```python
#2.View the series of different colors 

color_series
```

```python
#3.Create a series of three different car types and view it 

cars = ["Sedan","SUV","Hatchback"]

cars_series = pd.Series(cars)

print(cars_series)
```

```python
#4.Combine the Series of cars and colors into a Data Frame 

data = {"colors":color_series,"cars":cars_series}

df =pd.DataFrame(data)
df
```
